# Make Dataframe

```python
model = ['nb', 'logisticRegression','rfc','xgb','gbc']
label = ['toxic', 'stoxic','obscene','threat','insult','idHate','clean']
dataframe = {}
dataframe2 = {}

for nama in model:
    for item in label:
        dataframe['df{}_{}'.format(nama, item)] = pd.read_csv('{}_{}.csv'.format(nama, item))
        dataframe2['df{}_{}2'.format(nama, item)] = pd.read_csv('{}_{}2.csv'.format(nama, item))
```

```python
dfKey = list(dataframe.keys())
df2Key = list(dataframe2.keys())
```

```python
len(dfKey)
```

```python
dfResultToxic = pd.DataFrame

dataToxic = []
dataToxic2 = []
for i, j in zip(range(0,(len(dfKey)),7),range(0,(len(df2Key)),7)):
    dataToxic.append(dataframe[dfKey[i]])
    dataToxic2.append(dataframe2[df2Key[j]])

dfResultToxic = pd.concat(dataToxic).reset_index()
dfResultToxic2 = pd.concat(dataToxic2).reset_index()
```

```python
dfResultSToxic = pd.DataFrame

dataSToxic = []
dataSToxic2 = []
for i, j in zip(range(1,(len(dfKey)+1),7),range(1,(len(df2Key)+1),7)):
    dataSToxic.append(dataframe[dfKey[i]])
    dataSToxic2.append(dataframe2[df2Key[j]])

dfResultSToxic = pd.concat(dataSToxic).reset_index()
dfResultSToxic2 = pd.concat(dataSToxic2).reset_index()
```

```python
dfResultInsult = pd.DataFrame

dataInsult = []
dataInsult2 = []
for i, j in zip(range(4,(len(dfKey)+1),7),range(4,(len(df2Key)+1),7)):
    dataInsult.append(dataframe[dfKey[i]])
    dataInsult2.append(dataframe2[df2Key[j]])

dfResultInsult = pd.concat(dataInsult).reset_index()
dfResultInsult2 = pd.concat(dataInsult2).reset_index()
```

```python
dfResultIdHate = pd.DataFrame

dataIdHate = []
dataIdHate2 = []
for i, j in zip(range(5,(len(dfKey)+1),7),range(5,(len(df2Key)+1),7)):
    dataIdHate.append(dataframe[dfKey[i]])
    dataIdHate2.append(dataframe2[df2Key[j]])

dfResultIdHate = pd.concat(dataIdHate).reset_index()
dfResultIdHate2 = pd.concat(dataIdHate2).reset_index()
```

```python
dfResultClean = pd.DataFrame

dataClean = []
dataClean2 = []
for i, j in zip(range(6,(len(dfKey)+1),7),range(6,(len(df2Key)+1),7)):
    dataClean.append(dataframe[dfKey[i]])
    dataClean2.append(dataframe2[df2Key[j]])

dfResultClean = pd.concat(dataClean).reset_index()
dfResultClean2 = pd.concat(dataClean2).reset_index()
```
